```python
import os

files = os.listdir('csv')
print(files)
```

```python
import csv
readSetIds = set()

fieldNames = ['Num_Acc', 'date', 'temperature', 'dew_point', 'humidity', 'perspitation', 'wind', 'wind_speed','wind_gust','pressure','visibility', 'condition' ]
mergedFile = open('merged.csv', mode='w', newline='')
csvWriter = csv.DictWriter(mergedFile, fieldnames=fieldNames)
# csvWriter.writeheader()
totalNumberOfReads = 0

for fileName in files:
    with open(f'csv/{fileName}', mode='r') as f:
        csvFile = csv.DictReader(f, fieldnames=fieldNames)
        print(f'Reading from {fileName}')
        for index, line in enumerate(csvFile):
            if (index % 10000 == 0):
                print(f'Reading line {index}')
            if line['Num_Acc'] in readSetIds:
                continue
            csvWriter.writerow(line)
            readSetIds.add(line['Num_Acc'])

mergedFile.close()
```

```python
totalNumberOfReads = 0
for fileName in files:
    with open(f'csv/{fileName}', mode='r') as f:
        csvFile = csv.DictReader(f, fieldnames=fieldNames)
        totalNumberOfReads += sum(1 for row in csvFile)

print(f'This is the total number of reads in the scrapped data {totalNumberOfReads}')

totalNumberOfReads = 0
with open(f'merged.csv', mode='r') as f:
        csvFile = csv.DictReader(f, fieldnames=fieldNames)
        totalNumberOfReads += sum(1 for row in csvFile)

print(f'This is the total number of reads in the merged data {totalNumberOfReads}')

totalNumberOfReads = 0
with open(f'2022_23_pour_meteo.csv', mode='r') as f:
        csvFile = csv.DictReader(f, fieldnames=fieldNames)
        totalNumberOfReads += sum(1 for row in csvFile)

print(f'This is the total number of reads in the original csv {totalNumberOfReads}')
```

```python
mergedSet = set()
unFoundSet = set()
with open(f'merged.csv', mode='r') as f:
    csvFile = csv.DictReader(f, fieldnames=fieldNames)
    for index, line in enumerate(csvFile):
        mergedSet.add(line['Num_Acc'])

with open(f'2022_23_pour_meteo.csv', mode='r') as f:
    csvFile = csv.DictReader(f, delimiter=';')
    for index, line in enumerate(csvFile):
        if line['Num_Acc'] not in mergedSet:
            unFoundSet.add(line['Num_Acc'])

print(unFoundSet)


```

```python
totalNumberOfUnfoundData = 0
idsOfUnfoundDataSet = set()

with open(f'merged.csv', mode='r') as f:
    csvFile = csv.DictReader(f, fieldnames=fieldNames)
    for index, line in enumerate(csvFile):
        if line['temperature'] == 'inconnu':
            totalNumberOfUnfoundData += 1
            idsOfUnfoundDataSet.add(line['Num_Acc'])

print(f'Total number of unfound rows is {totalNumberOfUnfoundData}')
print(idsOfUnfoundDataSet)
```
